```python
import os
file_dir = "./output_dataframe/"
for i, file_name in enumerate(os.listdir(file_dir)):
    
    full_path = os.path.join(file_dir, file_name)
    
    if i == 0:
        output_df = pd.read_csv(full_path)
        
    else:
        df_joined = pd.read_csv(full_path)
        output_df = pd.concat([output_df, df_joined])
        
output_df
```

```python
# 重複を削除
final = output_df[(~output_df.duplicated())]

final
```

```python
final.to_csv("./output_dataframe/Japan_England_Germany_Spain_Italy.csv", index=False)
```
